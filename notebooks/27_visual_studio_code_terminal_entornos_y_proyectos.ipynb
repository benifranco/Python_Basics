## 1) Mapa mental: piezas de un proyecto

Cuando programamos en Python aparecen varias herramientas. Conviene separarlas:

| Pieza | Que es | Para que sirve |
|---|---|---|
| VS Code | Editor de codigo | Escribir, navegar y organizar archivos |
| Terminal | Interfaz de texto del sistema | Ejecutar comandos |
| Python | Interprete | Ejecutar archivos `.py` y notebooks |
| Proyecto | Carpeta principal | Reunir codigo, datos, pruebas y configuracion |
| Ambiente virtual | Python aislado para un proyecto | Evitar conflictos entre paquetes |
| `requirements.txt` | Lista de dependencias | Reproducir instalaciones |
| `.env` | Variables locales sensibles o configurables | Separar secretos y configuracion del codigo |

Una confusion comun es pensar que VS Code "es Python". No lo es: VS Code edita archivos; Python los ejecuta.

## 2) Visual Studio Code como editor

Visual Studio Code es un editor extensible. Para este curso nos interesan estas partes:

1. **Explorer**: muestra carpetas y archivos del proyecto.
2. **Editor**: zona donde escribimos codigo o Markdown.
3. **Terminal integrada**: permite ejecutar comandos sin salir de VS Code.
4. **Command Palette**: buscador de comandos de VS Code. Se abre con `Ctrl + Shift + P`.
5. **Source Control**: panel para trabajar con Git.
6. **Extensions**: panel para instalar herramientas adicionales.

Acciones frecuentes:

- Abrir una carpeta: `File > Open Folder`.
- Abrir terminal: `Terminal > New Terminal`.
- Seleccionar interprete de Python: `Ctrl + Shift + P` y luego `Python: Select Interpreter`.
- Ejecutar un archivo: abrir terminal y escribir `python archivo.py`.
- Abrir configuracion: `Ctrl + ,`.

Regla practica: abre siempre la carpeta raiz del proyecto, no solo un archivo suelto.

````python
import os
import platform
import sys
from pathlib import Path

print("Sistema operativo:", platform.system())
print("Version de Python:", sys.version.split()[0])
print("Interprete actual:", sys.executable)
print("Carpeta actual:", Path.cwd())
print("Ambiente virtual activo:", os.environ.get("VIRTUAL_ENV", "No detectado"))
````

La celda anterior no configura nada; solo muestra informacion del entorno que esta usando este notebook.

Si el interprete no corresponde al ambiente virtual esperado, en VS Code normalmente se corrige con:

1. `Ctrl + Shift + P`.
2. `Python: Select Interpreter`.
3. Elegir el Python que este dentro de `.venv`.

## 3) Terminal integrada y comandos basicos

La terminal permite hablar con el sistema operativo usando texto.

En Windows, VS Code suele abrir PowerShell. En macOS y Linux suele abrir Bash o Zsh.

### Comandos comunes

| Accion | PowerShell / Windows | macOS / Linux |
|---|---|---|
| Ver ubicacion actual | `pwd` | `pwd` |
| Listar archivos | `ls` o `dir` | `ls` |
| Cambiar de carpeta | `cd carpeta` | `cd carpeta` |
| Subir un nivel | `cd ..` | `cd ..` |
| Crear carpeta | `mkdir nombre` | `mkdir nombre` |
| Crear archivo vacio | `New-Item archivo.py` | `touch archivo.py` |
| Borrar archivo | `Remove-Item archivo.py` | `rm archivo.py` |
| Limpiar pantalla | `cls` | `clear` |
| Ejecutar script | `python archivo.py` | `python3 archivo.py` o `python archivo.py` |

Si una ruta tiene espacios, ponla entre comillas:

```powershell
cd "C:\Users\Tu Nombre\Desktop\mi proyecto"
```

En este curso conviene ejecutar comandos desde la carpeta raiz del proyecto.

### Comandos que se usan mucho en proyectos Python

```powershell
# Ver version de Python
python --version

# Ver version de pip
python -m pip --version

# Ejecutar un archivo
python main.py

# Instalar un paquete
python -m pip install pandas

# Ver paquetes instalados
python -m pip list
```

Usar `python -m pip` evita una confusion comun: instalar paquetes en un Python distinto al que ejecuta el proyecto.

## 4) Estructura recomendada de proyectos

Una estructura simple ayuda a que el proyecto sea entendible para ti y para otras personas.

Ejemplo:

```text
mi_proyecto/
  README.md
  requirements.txt
  .env
  .gitignore
  src/
    main.py
    funciones.py
  notebooks/
    exploracion.ipynb
  data/
    raw/
    processed/
  tests/
    test_funciones.py
```

Significado:

- `README.md`: explica que hace el proyecto y como ejecutarlo.
- `requirements.txt`: dependencias necesarias.
- `.env`: configuracion local y secretos. No se sube a Git.
- `.gitignore`: archivos o carpetas que Git debe ignorar.
- `src/`: codigo principal.
- `notebooks/`: exploracion, prototipos y explicaciones.
- `data/`: datos del proyecto. En proyectos reales se cuida si se suben o no.
- `tests/`: pruebas automatizadas.

No todos los proyectos necesitan todas las carpetas. La estructura debe ayudar, no estorbar.

````python
arbol = [
    ("", "mi_proyecto/"),
    ("  ", "README.md"),
    ("  ", "requirements.txt"),
    ("  ", ".env"),
    ("  ", ".gitignore"),
    ("  ", "src/"),
    ("    ", "main.py"),
    ("    ", "funciones.py"),
    ("  ", "notebooks/"),
    ("    ", "exploracion.ipynb"),
    ("  ", "data/"),
    ("    ", "raw/"),
    ("    ", "processed/"),
    ("  ", "tests/"),
    ("    ", "test_funciones.py"),
]

for indent, nombre in arbol:
    print(f"{indent}{nombre}")
````

## 5) Ambiente virtual con `venv`

Un ambiente virtual es una copia aislada de Python para un proyecto.

Sirve para evitar problemas como:

- Proyecto A necesita `pandas`.
- Proyecto B necesita otra version de `pandas`.
- Instalar todo globalmente puede romper cosas o volver dificil reproducir el trabajo.

La practica recomendada es crear un ambiente virtual por proyecto.

### Crear ambiente virtual en Windows PowerShell

```powershell
python -m venv .venv
.\.venv\Scripts\Activate.ps1
python -m pip install --upgrade pip
```

Si PowerShell bloquea la activacion, puede aparecer un error de politicas de ejecucion. En equipos personales se suele resolver con:

```powershell
Set-ExecutionPolicy -Scope CurrentUser RemoteSigned
```

En equipos institucionales puede requerir permiso del administrador.

### Crear ambiente virtual en macOS / Linux

```bash
python3 -m venv .venv
source .venv/bin/activate
python -m pip install --upgrade pip
```

### Desactivar el ambiente

```bash
deactivate
```

Cuando el ambiente esta activo, la terminal suele mostrar `(.venv)` al inicio de la linea.

### VS Code y el ambiente virtual

Despues de crear `.venv`, VS Code debe usar ese interprete.

Pasos:

1. Abrir la carpeta del proyecto.
2. Crear `.venv` desde la terminal.
3. Abrir `Ctrl + Shift + P`.
4. Buscar `Python: Select Interpreter`.
5. Elegir el interprete dentro de `.venv`.
6. Abrir una nueva terminal para que active el ambiente automaticamente.

Senales de que algo esta bien:

- `python --version` funciona.
- `python -m pip list` muestra paquetes del ambiente.
- `sys.executable` apunta a una ruta que contiene `.venv`.
- Al instalar un paquete, el notebook o script lo puede importar.

## 6) Paquetes, `pip` y `requirements.txt`

`pip` instala paquetes de Python.

Ejemplo:

```powershell
python -m pip install numpy pandas matplotlib
```

El archivo `requirements.txt` registra que paquetes necesita el proyecto.

Ejemplo simple:

```text
numpy
pandas
matplotlib
jupyterlab
```

Ejemplo con versiones minimas:

```text
numpy>=1.26
pandas>=2.0
matplotlib>=3.8
```

Ejemplo con versiones exactas:

```text
numpy==1.26.4
pandas==2.2.2
matplotlib==3.8.4
```

Versiones exactas ayudan a reproducir resultados, pero pueden ser mas rigidas. Para un curso introductorio, basta entender que `requirements.txt` permite instalar lo necesario en otra computadora.

### Comandos para `requirements.txt`

Crear desde el ambiente actual:

```powershell
python -m pip freeze > requirements.txt
```

Instalar desde el archivo:

```powershell
python -m pip install -r requirements.txt
```

Ver paquetes instalados:

```powershell
python -m pip list
```

Recomendacion: revisa el archivo antes de compartirlo. Si usaste un ambiente con muchos paquetes ajenos al proyecto, `pip freeze` puede incluir dependencias innecesarias.

````python
requirements_ejemplo = '''
numpy>=1.26
pandas>=2.0
matplotlib>=3.8
jupyterlab
'''.strip()

paquetes = []
for linea in requirements_ejemplo.splitlines():
    linea = linea.strip()
    if linea and not linea.startswith("#"):
        paquetes.append(linea)

print("Paquetes declarados:")
for paquete in paquetes:
    print("-", paquete)
````

## 7) Variables de entorno y archivo `.env`

Una variable de entorno guarda configuracion fuera del codigo.

Ejemplos:

- nombre de usuario,
- modo de ejecucion,
- ruta a una carpeta,
- llave de API,
- contrasena de base de datos.

El archivo `.env` suele verse asi:

```text
MODO=desarrollo
RUTA_DATOS=data/raw
API_KEY=valor_secreto
```

Reglas importantes:

1. No subas `.env` a GitHub.
2. Agrega `.env` a `.gitignore`.
3. No escribas secretos directamente en notebooks o scripts.
4. Puedes compartir un archivo `.env.example` sin secretos reales.

Ejemplo de `.env.example`:

```text
MODO=desarrollo
RUTA_DATOS=data/raw
API_KEY=pon_aqui_tu_api_key
```

````python
import os

# Esta celda no lee ningun archivo .env real.
# Solo muestra como consultar una variable de entorno desde Python.
modo = os.getenv("MODO", "desarrollo")
ruta_datos = os.getenv("RUTA_DATOS", "data/raw")

print("MODO:", modo)
print("RUTA_DATOS:", ruta_datos)
````

### Leer `.env` con `python-dotenv`

Python no lee archivos `.env` automaticamente. Una opcion comun es instalar `python-dotenv`.

```powershell
python -m pip install python-dotenv
```

Uso tipico:

```python
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv("API_KEY")
```

Si usas `python-dotenv`, recuerda agregarlo a `requirements.txt`.

## 8) Extensiones utiles para VS Code

No necesitas instalar decenas de extensiones. Para empezar, estas son suficientes:

| Extension | Utilidad |
|---|---|
| Python | Soporte principal para Python en VS Code |
| Pylance | Autocompletado, analisis y ayuda con tipos |
| Jupyter | Abrir y ejecutar notebooks `.ipynb` |
| Ruff | Revisar estilo, errores comunes y ordenar imports |
| Black Formatter | Formatear codigo Python automaticamente |
| GitLens | Ver historial y contexto de cambios Git |
| GitHub Pull Requests | Trabajar con issues y pull requests desde VS Code |
| Rainbow CSV | Leer archivos CSV con columnas coloreadas |
| Markdown All in One | Escribir documentacion Markdown con mas comodidad |

Recomendacion para principiantes:

1. Instala `Python`, `Pylance` y `Jupyter`.
2. Cuando ya escribas scripts con frecuencia, agrega `Ruff` y `Black Formatter`.
3. Si trabajas con GitHub, agrega `GitLens` y `GitHub Pull Requests`.

Evita instalar extensiones sin entender para que sirven: demasiadas extensiones pueden volver lento o confuso el editor.

### Configuraciones utiles de VS Code

Puedes activar estas opciones desde la interfaz grafica de configuracion:

- `Format On Save`: formatea al guardar.
- `Trim Trailing Whitespace`: elimina espacios sobrantes.
- `Auto Save`: guarda automaticamente, si te resulta comodo.
- `Default Formatter`: selecciona Black Formatter si lo instalaste.

Un archivo `.vscode/settings.json` puede guardar configuraciones del proyecto.

Ejemplo:

```json
{
  "editor.formatOnSave": true,
  "editor.trimAutoWhitespace": true,
  "python.analysis.typeCheckingMode": "basic"
}
```

Evita guardar rutas absolutas de tu computadora en archivos compartidos, porque no funcionaran igual para tus companeros.

## 9) Flujo recomendado de trabajo

Un flujo ordenado para empezar un proyecto seria:

1. Crear carpeta del proyecto.
2. Abrir esa carpeta en VS Code.
3. Crear ambiente virtual `.venv`.
4. Activar el ambiente.
5. Seleccionar el interprete correcto en VS Code.
6. Crear estructura minima: `src/`, `notebooks/`, `data/`, `README.md`.
7. Instalar paquetes necesarios.
8. Registrar dependencias en `requirements.txt`.
9. Crear `.gitignore` e ignorar `.venv`, `.env`, caches y archivos temporales.
10. Escribir codigo en pequenos pasos y ejecutar desde la terminal.
11. Documentar como reproducir el proyecto.

Ejemplo de `.gitignore` para Python:

```text
.venv/
.env
__pycache__/
.ipynb_checkpoints/
*.pyc
```

El objetivo es que otra persona pueda abrir el proyecto, crear su ambiente, instalar dependencias y ejecutar el codigo.

## 10) Checklist de diagnostico

Cuando algo no funciona, revisa en este orden:

1. Estoy en la carpeta correcta? Usa `pwd`.
2. El ambiente virtual esta activo?
3. VS Code esta usando el interprete correcto?
4. Instale el paquete dentro del ambiente correcto?
5. El archivo que intento ejecutar existe en esta carpeta?
6. La ruta tiene espacios y necesita comillas?
7. El paquete esta registrado en `requirements.txt`?
8. Estoy exponiendo algun secreto en el codigo?

Muchos errores de principiantes no son de Python, sino de entorno, rutas o interpretes distintos.
