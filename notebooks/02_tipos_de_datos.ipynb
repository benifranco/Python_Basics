## Â¿Por quÃ© importan los tipos de datos?

Los tipos de datos son la **base de cualquier programa**. Python necesita saber si estÃ¡s trabajando con nÃºmeros, texto o valores verdadero/falso para realizar operaciones correctas.

## 1. Tipos de Datos Primitivos

### 1.1 INT (Enteros)

**DefiniciÃ³n**: NÃºmeros enteros sin decimales.

**CaracterÃ­sticas**:
- Rango ilimitado (puedes tener nÃºmeros muy grandes)
- Operaciones: suma (+), resta (-), multiplicaciÃ³n (*), divisiÃ³n entera (//), mÃ³dulo (%)
- No tienen punto decimal

```python
# Creando variables de tipo int
edad = 25
cantidad = 100
temperatura = -5

print(f"Edad: {edad}")
print(f"Cantidad: {cantidad}")
print(f"Temperatura: {temperatura}Â°C")

# Operaciones con enteros
suma = 10 + 5
resta = 10 - 3
multiplicacion = 4 * 7
division_entera = 15 // 2  # Resultado sin decimales
modulo = 15 % 2  # Resto de la divisiÃ³n

print(f"\nOperaciones:")
print(f"10 + 5 = {suma}")
print(f"10 - 3 = {resta}")
print(f"4 * 7 = {multiplicacion}")
print(f"15 // 2 = {division_entera}  (divisiÃ³n entera)")
print(f"15 % 2 = {modulo}  (mÃ³dulo - resto)")
```

### 1.2 FLOAT (Decimales)

**DefiniciÃ³n**: NÃºmeros con punto decimal.

**CaracterÃ­sticas**:
- Representan valores decimales
- PrecisiÃ³n limitada (cuidado con comparaciones exactas)
- Operaciones similares a int, pero con decimales

```python
# Creando variables de tipo float
precio = 19.99
altura = 1.75  # metros
pi = 3.14159
temperatura = 36.5  # grados

print(f"Precio: ${precio}")
print(f"Altura: {altura} m")
print(f"Pi â‰ˆ {pi}")
print(f"Temperatura: {temperatura}Â°C")

# Operaciones con floats
area_circulo = pi * (5 ** 2)  # pi * radio^2
promedio = (85.5 + 90.3 + 88.7) / 3

print(f"\nOperaciones:")
print(f"Ãrea de cÃ­rculo con radio 5: {area_circulo:.2f}")
print(f"Promedio de calificaciones: {promedio:.2f}")
```

### 1.3 STR (Cadenas de Texto)

**DefiniciÃ³n**: Secuencia de caracteres (texto).

**CaracterÃ­sticas**:
- Se escriben entre comillas simples '' o dobles ""
- Inmutables (no se pueden cambiar despuÃ©s de creadas)
- Tienen mÃ©todos Ãºtiles

```python
# Creando variables de tipo str
nombre = "Juan"
apellido = 'PÃ©rez'
mensaje = "Hola, bienvenido a Python"
email = "juan@ejemplo.com"

print(f"Nombre: {nombre}")
print(f"Apellido: {apellido}")
print(f"Mensaje: {mensaje}")
print(f"Email: {email}")

# MÃ©todos Ãºtiles de strings
print(f"\nMÃ©todos de strings:")
print(f"MayÃºsculas: {nombre.upper()}")
print(f"MinÃºsculas: {nombre.lower()}")
print(f"Largo: {len(mensaje)} caracteres")
print(f"Reemplazar: {mensaje.replace('Python', 'JavaScript')}")
print(f"Â¿Contiene 'bienvenido'?: {'bienvenido' in mensaje}")
```

### 1.4 BOOL (Booleanos)

**DefiniciÃ³n**: Solo dos valores posibles: `True` (verdadero) o `False` (falso).

**CaracterÃ­sticas**:
- Resultado de comparaciones y operadores lÃ³gicos
- Fundamentales en condicionales

```python
# Creando variables de tipo bool
es_estudiante = True
tiene_licencia = False

print(f"Â¿Es estudiante?: {es_estudiante}")
print(f"Â¿Tiene licencia?: {tiene_licencia}")

# Booleans de comparaciones
print(f"\nResultados de comparaciones:")
print(f"5 > 3 = {5 > 3}")
print(f"10 < 5 = {10 < 5}")
print(f"7 == 7 = {7 == 7}")
print(f"8 != 3 = {8 != 3}")

# Operadores lÃ³gicos
print(f"\nOperadores lÃ³gicos:")
print(f"True and False = {True and False}")
print(f"True or False = {True or False}")
print(f"not True = {not True}")
```

### 1.5 Identificar tipos con type()

Usa la funciÃ³n `type()` para saber quÃ© tipo de dato tienes.

```python
# Usando type() para identificar tipos
numero_entero = 42
numero_decimal = 3.14
texto = "Python"
booleano = True

print(f"type(42) = {type(numero_entero)}")
print(f"type(3.14) = {type(numero_decimal)}")
print(f"type('Python') = {type(texto)}")
print(f"type(True) = {type(booleano)}")

# Comparar tipos
print(f"\nÂ¿42 es int?: {type(numero_entero) == int}")
print(f"Â¿'Python' es str?: {type(texto) == str}")
```

### 2.2 F-Strings (Formato de Cadenas)

F-strings permiten insertar variables dentro de texto de forma legible.

```python
# F-strings - La forma moderna y clara
nombre = "MarÃ­a"
edad = 28
ciudad = "MÃ©xico"

print(f"Mi nombre es {nombre}")
print(f"Tengo {edad} aÃ±os")
print(f"Vivo en {ciudad}")

# Expresiones dentro de f-strings
a = 10
b = 5
print(f"\n{a} + {b} = {a + b}")
print(f"{a} * {b} = {a * b}")

# Formateo de decimales
precio = 19.99
print(f"\nPrecio con 2 decimales: ${precio:.2f}")

numero = 42
print(f"NÃºmero centrado: |{numero:^10}|")
```

## 3. ConversiÃ³n de Tipos (Casting)

Frecuentemente necesitas convertir un tipo de dato a otro. Usa `int()`, `float()`, `str()` o `bool()`.

### 3.1 Convertir a INT

```python
# Convertir str a int
numero_texto = "42"
numero_entero = int(numero_texto)
print(f"'{numero_texto}' convertido a int: {numero_entero}")
print(f"Tipo: {type(numero_entero)}")

# Convertir float a int (trunca decimales)
numero_decimal = 3.99
numero_entero = int(numero_decimal)
print(f"\n{numero_decimal} convertido a int: {numero_entero}")
print(f"Nota: Se pierden los decimales")

# ConversiÃ³n Ãºtil: texto de input
# cantidad = int(input("Â¿CuÃ¡ntos artÃ­culos?: "))
cantidad = 5
print(f"\nCantidad de artÃ­culos: {cantidad} (tipo: {type(cantidad).__name__})")
```

### 3.2 Convertir a FLOAT

```python
# Convertir str a float
valor_texto = "3.14"
valor_decimal = float(valor_texto)
print(f"'{valor_texto}' convertido a float: {valor_decimal}")
print(f"Tipo: {type(valor_decimal)}")

# Convertir int a float
numero_int = 5
numero_float = float(numero_int)
print(f"\n{numero_int} convertido a float: {numero_float}")
print(f"Nota: Se agrega .0")

# Uso prÃ¡ctico
# precio = float(input("Ingresa el precio: $"))
precio = 19.99
total = precio * 3
print(f"\nPrecio unitario: ${precio:.2f}")
print(f"Total (3 unidades): ${total:.2f}")
```

### 3.3 Convertir a STR

```python
# Convertir int a str
numero = 42
numero_texto = str(numero)
print(f"{numero} convertido a str: '{numero_texto}'")
print(f"Tipo: {type(numero_texto)}")

# Convertir float a str
pi = 3.14159
pi_texto = str(pi)
print(f"\n{pi} convertido a str: '{pi_texto}'")

# Convertir bool a str
booleano = True
booleano_texto = str(booleano)
print(f"\n{booleano} convertido a str: '{booleano_texto}'")

# Concatenar requiere todo como str
print(f"\nConcatenaciÃ³n requiere strings:")
print("El nÃºmero es " + str(42) + " y es importante")
```

### 3.4 Convertir a BOOL

```python
# Valores "falsy" (se convierten a False)
print("Conversiones a bool:")
print(f"bool(0) = {bool(0)}")
print(f"bool('') = {bool('')}")
print(f"bool([]) = {bool([])}")
print(f"bool(None) = {bool(None)}")

# Valores "truthy" (se convierten a True)
print(f"\nbool(1) = {bool(1)}")
print(f"bool(42) = {bool(42)}")
print(f"bool('Texto') = {bool('Texto')}")
print(f"bool([1, 2, 3]) = {bool([1, 2, 3])}")
```

## 4. Errores Comunes y CÃ³mo Evitarlos

### 4.1 TypeError - OperaciÃ³n entre tipos incompatibles

```python
# âŒ ERROR: Sumar string y nÃºmero
try:
    resultado = "5" + 5
except TypeError as e:
    print(f"âŒ ERROR TypeError: {e}")
    print(f"Causa: No puedes sumar un string con un nÃºmero\n")

# âœ… SOLUCIÃ“N: Convertir al mismo tipo
resultado = int("5") + 5
print(f"âœ… CORRECTO: int('5') + 5 = {resultado}")

# âŒ ERROR: Multiplicar nÃºmero por string
try:
    resultado = 5 * "texto"
except TypeError as e:
    print(f"\nâŒ ERROR TypeError: {e}")
```

### 4.2 ValueError - ConversiÃ³n invÃ¡lida

```python
# âŒ ERROR: Convertir texto que no es nÃºmero
try:
    numero = int("abc")
except ValueError as e:
    print(f"âŒ ERROR ValueError: {e}")
    print(f"Causa: 'abc' no es un nÃºmero vÃ¡lido\n")

# âœ… SOLUCIÃ“N: Validar antes de convertir
numero_texto = "abc"
if numero_texto.isdigit():
    numero = int(numero_texto)
    print(f"âœ… NÃºmero convertido: {numero}")
else:
    print(f"âœ… El texto '{numero_texto}' no es un nÃºmero vÃ¡lido")

# âŒ ERROR: Convertir float con punto a int
try:
    numero = int("3.14")
except ValueError as e:
    print(f"\nâŒ ERROR ValueError: {e}")
    print(f"Causa: int() no acepta decimales como string")
    print(f"SoluciÃ³n: usar float() primero")
    numero = int(float("3.14"))
    print(f"int(float('3.14')) = {numero}")
```
