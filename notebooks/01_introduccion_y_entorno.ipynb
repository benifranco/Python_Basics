## 3Ô∏è‚É£ Entornos Virtuales con venv

### ¬øPor qu√© necesitamos entornos virtuales?

Un **entorno virtual** es una carpeta aislada donde instalamos las librer√≠as espec√≠ficas de un proyecto.

**Ventajas:**
- ‚úÖ Aislar dependencias por proyecto
- ‚úÖ Evitar conflictos de versiones
- ‚úÖ Facilitar colaboraci√≥n entre desarrolladores
- ‚úÖ Proyecto port√°til y reproducible

### Crear un Entorno Virtual

#### En Windows (PowerShell o CMD):

```bash
# Crear el entorno virtual
python -m venv env

# Activar el entorno
env\Scripts\activate

# Deber√≠as ver (env) al inicio de la l√≠nea
# (env) C:\Users\usuario\proyecto>
```

#### En macOS y Linux:

```bash
# Crear el entorno virtual
python3 -m venv env

# Activar el entorno
source env/bin/activate

# Ver√°s (env) al inicio
# (env) usuario@computadora:~/proyecto$
```

### ¬øC√≥mo s√© que est√° activado?

Cuando veas `(env)` al inicio de tu l√≠nea de comando, el entorno est√° activado:

```
(env) usuario@computadora:~/proyecto$
```

### Desactivar el entorno

```bash
deactivate
```

## 4Ô∏è‚É£ pip: Gestor de Paquetes

### ¬øQu√© es pip?

`pip` es el **gestor de paquetes** de Python. Te permite instalar, actualizar y desinstalar librer√≠as (tambi√©n llamadas m√≥dulos o paquetes).

### Comandos B√°sicos de pip

```bash
# Ver versi√≥n de pip
pip --version
# pip 22.3.1 from /Users/usuario/env/lib/python3.11/site-packages/pip

# Instalar una librer√≠a
pip install requests

# Instalar una versi√≥n espec√≠fica
pip install numpy==1.21.0

# Instalar m√∫ltiples librer√≠as
pip install flask pandas matplotlib

# Actualizar una librer√≠a
pip install --upgrade numpy

# Desinstalar una librer√≠a
pip uninstall requests

# Ver librer√≠as instaladas
pip list

# Generar requirements.txt (lista de dependencias)
pip freeze > requirements.txt

# Instalar desde requirements.txt
pip install -r requirements.txt
```

### Archivo requirements.txt

Es un archivo de texto que lista todas las dependencias de tu proyecto:

```txt
requests==2.28.1
pandas==1.5.2
matplotlib==3.6.2
flask==2.2.2
```

**Beneficio:** Para recrear el entorno en otra m√°quina:
```bash
pip install -r requirements.txt
```

## 9Ô∏è‚É£ Errores Comunes y Soluciones

### ‚ùå "python: command not found"
**Problema**: Python no est√° en PATH

**Soluci√≥n**:
- **Windows**: Reinstala Python marcando **"Add Python to PATH"**
- **macOS**: Usa `python3` en lugar de `python`
- **Linux**: Instala Python con `sudo apt install python3`

---

### ‚ùå "ModuleNotFoundError: No module named 'requests'"
**Problema**: La librer√≠a no est√° instalada en el entorno actual

**Soluci√≥n**:
1. Verifica que el entorno virtual est√° activado (deber√≠as ver `(env)` en tu terminal)
2. Instala la librer√≠a:
   ```bash
   pip install requests
   ```

---

### ‚ùå "No such file or directory: hola_mundo.py"
**Problema**: El archivo no existe o est√°s en la carpeta incorrecta

**Soluci√≥n**:
```bash
# Verifica d√≥nde est√°s
pwd

# Lista los archivos
ls

# Aseg√∫rate de tener hola_mundo.py en esta carpeta
```

---

### ‚ùå "The virtual environment was not activated"
**Problema**: Olvidaste activar el entorno virtual

**Soluci√≥n**: Act√≠valo seg√∫n tu sistema operativo
```bash
# Windows
env\Scripts\activate

# macOS/Linux
source env/bin/activate
```
