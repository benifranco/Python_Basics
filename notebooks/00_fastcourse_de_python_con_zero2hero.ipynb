## 6) Diccionarios

Un diccionario guarda pares `clave: valor`.

````python
alumno = {
    "nombre": "Sofia",
    "edad": 19,
    "promedio": 9.1
}

print(alumno)
print("Nombre:", alumno["nombre"])

alumno["edad"] = 20
alumno["carrera"] = "Ingenieria"
print("Actualizado:", alumno)

for clave, valor in alumno.items():
    print(clave, "->", valor)
````

## 8) Ciclos (`for` y `while`)

Un ciclo repite acciones para evitar codigo duplicado.

````python
print("Ciclo for con range:")
for i in range(1, 6):
    print("i =", i)

print("\nCiclo while:")
contador = 1
while contador <= 3:
    print("contador =", contador)
    contador += 1
````

## 10) Programacion Orientada a Objetos (POO)

POO organiza el codigo en clases y objetos.

- Clase: plantilla.
- Objeto: instancia concreta de la clase.
- Atributos: datos del objeto.
- Metodos: acciones del objeto.

````python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentarse(self):
        return f"Hola, soy {self.nombre} y tengo {self.edad} anos"

persona1 = Persona("Elena", 22)
print(persona1.presentarse())

persona2 = Persona("Mario", 19)
print(persona2.presentarse())
````
