### Teoria 2

## 2. print() a fondo

`print()` tiene parametros utiles:

- `sep`: separador entre valores (por defecto es espacio).
- `end`: lo que se imprime al final (por defecto es salto de linea).
- `flush`: fuerza a mostrar la salida de inmediato.

## Extra de probabilidad y estadistica

Reporte estadistico formateado para comunicacion de resultados.


```r
x <- c(7.2,8.1,9.0,6.8,8.4)
cat(sprintf('media=%.2f mediana=%.2f sd=%.2f\n', mean(x), median(x), sd(x)))
```
