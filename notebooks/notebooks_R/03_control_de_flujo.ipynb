## Operadores relacionales y logicos

El control de flujo depende de expresiones booleanas (`TRUE/FALSE`).
Operadores clave:

- Relacionales: `>`, `<`, `>=`, `<=`, `==`, `!=`
- Logicos: `&`, `|`, `!`


```r
# Define dos valores para comparar.
a <- 12
b <- 7

# Operadores relacionales.
a > b
a == b
a != b

# Operadores logicos compuestos.
(a > 10) & (b > 5)
(a > 15) | (b > 5)
!(a > b)
```

## Condicional basico: if / else if / else

Este patron se usa para clasificar o decidir rutas de ejecucion.


```r
# Clasifica una nota academica.
nota <- 8.3

if (nota >= 9) {
  categoria <- 'excelente'
} else if (nota >= 7) {
  categoria <- 'aprobado'
} else {
  categoria <- 'reprobado'
}

categoria
```

## Condiciones anidadas y legibilidad

Puedes anidar condiciones, pero evita anidacion excesiva.
Una buena practica es usar reglas claras de mayor a menor prioridad.


```r
# Ejemplo: clasifica edad y situacion laboral.
edad <- 23
trabaja <- TRUE

if (edad < 18) {
  perfil <- 'menor'
} else {
  if (trabaja) {
    perfil <- 'adulto con empleo'
  } else {
    perfil <- 'adulto sin empleo'
  }
}

perfil
```

## `if` escalar vs `ifelse` vectorizado

- `if` espera una sola condicion (`TRUE/FALSE`).
- `ifelse` evalua un vector y regresa un vector de resultados.


```r
# Vector de edades.
edades <- c(12, 17, 21, 35)

# ifelse aplica condicion elemento por elemento.
grupo <- ifelse(edades < 18, 'menor', 'adulto')

edades
grupo

# Ejemplo adicional: etiqueta de riesgo por presion arterial.
presion <- c(110, 130, 145, 118)
riesgo <- ifelse(presion >= 140, 'alto', 'control')
riesgo
```

## `switch` para menus de opciones

`switch` es util cuando eliges entre varias acciones segun una etiqueta.


```r
# Opcion elegida por el usuario.
opcion <- 'media'

# Menu de operaciones.
resultado <- switch(
  opcion,
  suma = sum(1:10),
  media = mean(1:10),
  maximo = max(1:10),
  'opcion no valida'
)

resultado
```

## Validacion de entradas (casos borde)

Control de flujo tambien sirve para proteger funciones contra entradas invalidas.


```r
# Funcion con validacion de rango.
clasificar_prob <- function(p) {
  # Rechaza valores fuera de [0,1].
  if (p < 0 || p > 1) {
    stop('p debe estar entre 0 y 1')
  }

  # Clasificacion por intervalos.
  if (p < 0.2) {
    return('baja')
  } else if (p < 0.7) {
    return('media')
  } else {
    return('alta')
  }
}

clasificar_prob(0.15)
clasificar_prob(0.55)
clasificar_prob(0.91)
```

## Mini caso aplicado: decision de beca

Construimos una regla simple con varias condiciones.


```r
# Datos de ejemplo.
promedio <- 8.9
ingreso_familiar <- 6200
faltas <- 2

# Regla de beca: promedio alto, ingreso bajo y pocas faltas.
if (promedio >= 8.5 && ingreso_familiar <= 7000 && faltas <= 3) {
  decision <- 'beca aprobada'
} else {
  decision <- 'beca rechazada'
}

decision
```

## Extra de probabilidad y estadistica

Vamos a estimar una probabilidad condicional empirica con simulacion:

- Evento A: aprobar examen.
- Evento B: estudiar mas de 2 horas.

Objetivo: estimar `P(A | B)`.


```r
# Semilla para reproducibilidad.
set.seed(123)

# Simula horas de estudio (discreta simple).
horas <- sample(0:5, size = 5000, replace = TRUE,
                prob = c(0.10, 0.15, 0.20, 0.25, 0.20, 0.10))

# Modelo simple: mayor probabilidad de aprobar con mas horas.
p_aprueba <- ifelse(horas <= 1, 0.25,
                    ifelse(horas <= 3, 0.60, 0.82))

# Simula aprobacion como Bernoulli segun p_aprueba.
aprueba <- rbinom(n = length(horas), size = 1, prob = p_aprueba)

# Define eventos logicos.
A <- aprueba == 1
B <- horas > 2

# Estimaciones empiricas.
P_A <- mean(A)
P_B <- mean(B)
P_AyB <- mean(A & B)
P_A_dado_B <- P_AyB / P_B

cat('P(A) =', round(P_A, 4), '
')
cat('P(B) =', round(P_B, 4), '
')
cat('P(A y B) =', round(P_AyB, 4), '
')
cat('P(A | B) =', round(P_A_dado_B, 4), '
')
```

## Analogias rapidas R y Python

- `if/elif/else` en Python equivale a `if/else if/else` en R.
- `ifelse` vectorizado en R cumple un papel parecido a operaciones vectorizadas con `numpy.where`.
- `switch` en R es una alternativa compacta para menus simples.


```r
# Ejemplo comparativo: vectorizacion condicional en R.
# En Python se resolveria con list comprehension o numpy.where.
x <- c(3, 7, 1, 9, 5)
etiqueta <- ifelse(x >= 5, 'alto', 'bajo')

data.frame(x = x, etiqueta = etiqueta)
```

## Errores comunes de esta sesion

1. Usar `if` con un vector completo en vez de `ifelse`.
2. Escribir condiciones ambiguas sin parentesis.
3. No validar entradas antes de aplicar reglas.
4. Crear demasiada anidacion y perder legibilidad.


```r
# Error comun: usar if con vector.
valores <- c(2, 5, 8)

# Esto da warning/error conceptual porque if espera un unico TRUE/FALSE.
# if (valores > 4) {
#   print('hay valores altos')
# }

# Forma correcta para vector: any/all o ifelse.
if (any(valores > 4)) {
  print('Al menos un valor es mayor que 4')
}

if (all(valores > 4)) {
  print('Todos son mayores que 4')
} else {
  print('No todos son mayores que 4')
}
```
