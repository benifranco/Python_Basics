## Variables y asignacion

En R, la asignacion tipica se hace con `<-`.
Tambien puedes usar `=`, pero en practica de analisis se prefiere `<-` por claridad.


```r
# Asignacion con operador recomendado en R.
x <- 10
y <- 3

# Operaciones aritmeticas basicas.
suma <- x + y
producto <- x * y
potencia <- x^2

# Mostrar resultados.
cat('suma =', suma, '
')
cat('producto =', producto, '
')
cat('potencia =', potencia, '
')
```

## Tipos de datos basicos

R tiene tipos atomicos principales: numerico, entero, caracter y logico.
Entender tipos evita errores cuando combines datos.


```r
# Ejemplos de tipos atomicos.
v_num <- c(1.2, 2.5, 3.8)
v_int <- c(1L, 2L, 3L)
v_chr <- c('a', 'b', 'c')
v_log <- c(TRUE, FALSE, TRUE)

# typeof indica el tipo interno.
typeof(v_num)
typeof(v_int)
typeof(v_chr)
typeof(v_log)

# str resume estructura de un objeto.
str(v_num)
```

## Vectores y operaciones vectorizadas

Una fortaleza de R es trabajar con vectores completos sin bucles explicitos.
Esto hace codigo mas conciso y, muchas veces, mas rapido.


```r
# Crea un vector numerico.
notas <- c(8.5, 9.0, 7.8, 10.0, 8.9)

# Operacion vectorizada: sumar 0.5 a todas las notas.
notas_ajustadas <- notas + 0.5

# Comparaciones vectorizadas devuelven un vector logico.
aprobado <- notas >= 8

notas
notas_ajustadas
aprobado
```

## Primer vistazo estadistico

Con un vector numerico puedes calcular estadisticos descriptivos basicos:

- media,
- mediana,
- desviacion estandar,
- minimo y maximo.


```r
# Estadistica descriptiva basica.
cat('media =', mean(notas), '
')
cat('mediana =', median(notas), '
')
cat('sd =', sd(notas), '
')
cat('min =', min(notas), '
')
cat('max =', max(notas), '
')

# summary ofrece un resumen rapido.
summary(notas)
```

## Condicional basico

El control de flujo permite tomar decisiones segun condiciones.
En esta sesion solo veremos el patron minimo con `if/else`.


```r
# Clasifica una calificacion usando if/else.
calificacion <- 8.3

if (calificacion >= 9) {
  resultado <- 'excelente'
} else if (calificacion >= 7) {
  resultado <- 'aprobado'
} else {
  resultado <- 'reprobado'
}

resultado
```

## Primera grafica en R

Visualizar datos es parte central de ciencia de datos.
Aqui graficamos una funcion matematica simple para practicar.


```r
# Construye puntos del eje x.
u <- seq(0, 2*pi, length.out = 300)

# Calcula seno en cada punto.
v <- sin(u)

# Grafica de linea con etiquetas.
plot(u, v,
     type = 'l',
     col = 'steelblue',
     lwd = 2,
     main = 'Primera grafica en R',
     xlab = 'x',
     ylab = 'sin(x)')

grid()
```

## Extra de probabilidad y estadistica: simulacion Bernoulli

Una variable Bernoulli toma valores `0` o `1`.
Podemos simularla y comparar la frecuencia relativa con la probabilidad teorica.


```r
# Fija semilla para reproducibilidad.
set.seed(2026)

# Simula 1000 ensayos Bernoulli con p = 0.30.
# sample toma 0/1 con probabilidades dadas.
n <- 1000
x <- sample(c(0, 1), size = n, replace = TRUE, prob = c(0.70, 0.30))

# Frecuencia relativa de 1 (estimador empirico de p).
p_hat <- mean(x)

# Resumen de resultados.
cat('p teorica = 0.30
')
cat('p estimada =', round(p_hat, 4), '
')
cat('sd muestral =', round(sd(x), 4), '
')

# Visualiza conteos de 0 y 1.
barplot(table(x), col = c('gray70', 'tomato'), main = 'Conteo Bernoulli simulada', xlab = 'valor', ylab = 'frecuencia')
```

## Analogias rapidas R y Python

- `print()` existe en ambos lenguajes.
- `vector` en R se parece al uso de arreglos/listas en Python, pero es mas central en R base.
- `NA` en R juega un papel similar a `None/NaN`, pero su manejo en operaciones estadisticas es muy explicito.


```r
# Ejemplo de analogia: operacion vectorizada en R.
# En Python puro suele hacerse con bucle; en R se aplica directo al vector.

v <- c(1, 2, 3, 4, 5)
v_doblado <- v * 2
v_doblado

# Manejo de faltantes: NA en R.
w <- c(10, 12, NA, 14)
mean(w)                 # devuelve NA
mean(w, na.rm = TRUE)   # ignora faltantes de forma explicita
```

## Errores comunes en esta sesion

1. No ejecutar celdas en orden y usar variables no definidas.
2. Mezclar tipos sin revisar coercion implicita.
3. Olvidar `set.seed()` en simulaciones y no poder reproducir resultados.
4. Calcular estadisticos con datos faltantes sin `na.rm = TRUE`.


```r
# Ejemplo de error comun 1: no fijar semilla.
# Cada corrida da resultados distintos.
x1 <- sample(c(0, 1), size = 10, replace = TRUE)
x2 <- sample(c(0, 1), size = 10, replace = TRUE)
x1
x2

# Correccion: fijar semilla para reproducibilidad.
set.seed(123)
y1 <- sample(c(0, 1), size = 10, replace = TRUE)
set.seed(123)
y2 <- sample(c(0, 1), size = 10, replace = TRUE)
y1
y2
```
