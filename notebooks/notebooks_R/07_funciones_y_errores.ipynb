## 1. ¿Qué es una función?

Una función es un bloque reutilizable de código que recibe entradas (parámetros)
y devuelve una salida. Ayuda a evitar repetición, mejorar la lectura y probar
partes específicas del programa.

### Teoria 2

## 2. Parámetros y argumentos

Puedes pasar valores por posición o por nombre. Lo importante es respetar el orden
y los nombres definidos en la función.

### Teoria 3

## 3. Valores por defecto

Los parámetros opcionales permiten llamadas más simples y evitan repetir valores
comunes.


```r
area_rect <- function(base, altura=1){if(base<0 || altura<0) stop('base y altura no negativas'); base*altura}
area_rect(4,2); area_rect(5)
```

### Bloque teorico 2

### Teoria 4

## 4. `return` y múltiples valores

Una función puede devolver más de un valor. En R esto se representa como
una tupla.

### Teoria 5

## 5. Docstrings y ayuda rápida

Las docstrings describen lo que hace la función, sus parámetros y su salida.
Puedes consultarlas con `.__doc__`.

### Teoria 6

## 6. Alcance (scope): variables locales y globales

Las variables dentro de una función son locales. Una variable global no cambia
a menos que se declare con `global` (no recomendado en cursos básicos).


```r
resumen_num <- function(x, ...) c(media=mean(x,...), sd=sd(x,...), min=min(x,...), max=max(x,...))
resumen_num(c(1,2,NA,5), na.rm=TRUE)
```

### Bloque teorico 3

### Teoria 7

## 7. Manejo de errores con `try/except/else/finally`

`try` permite ejecutar un bloque que podría fallar. `except` captura el error,
`else` se ejecuta si no hay error y `finally` siempre se ejecuta.

### Teoria 8

## 8. `raise` para validar datos

Cuando los datos son inválidos, puedes lanzar un error para detener y avisar
claramente.

### Teoria 9

## 10. Resumen de Conceptos Clave

| Concepto | Qué es | Ejemplo |
|----------|--------|---------|
| `def` | Define una función | `function suma(a, b):` |
| Parámetros | Variables de entrada | `function f(x):` |
| Argumentos | Valores pasados | `f(10)` |
| `return` | Valor devuelto | `return total` |
| Docstring | Documentación | `"""..."""` |
| `try/except` | Manejo de errores | `try: ... except ValueError:` |
| `raise` | Lanza un error | `raise ValueError("msg")` |

## Extra de probabilidad y estadistica

Bootstrap de la media para cuantificar incertidumbre.


```r
set.seed(42)
x <- rnorm(60,10,2)
boot_means <- replicate(2000, mean(sample(x, replace=TRUE)))
quantile(boot_means, c(0.025,0.5,0.975))
```
