## Tipos atomicos en R

Los tipos basicos que vas a usar constantemente son:

- `numeric` (double),
- `integer`,
- `character`,
- `logical`.


```r
# Crea un vector de cada tipo atomico.
v_num <- c(1.2, 2.5, 3.8)
v_int <- c(1L, 2L, 3L)
v_chr <- c('a', 'b', 'c')
v_log <- c(TRUE, FALSE, TRUE)

# typeof muestra el tipo interno del objeto.
typeof(v_num)
typeof(v_int)
typeof(v_chr)
typeof(v_log)

# class da una vista de clase (a veces coincide, a veces no).
class(v_num)
class(v_int)
class(v_chr)
class(v_log)
```

## Coercion implicita

Cuando mezclas tipos en un mismo vector, R intenta convertir todo a un tipo comun.
Esto puede ser util, pero tambien causar errores silenciosos.


```r
# Mezcla de tipos en un solo vector.
mezcla_1 <- c(1, TRUE)
mezcla_2 <- c(1, 'texto')
mezcla_3 <- c(TRUE, 'texto')

# Observa como cambia el tipo final.
mezcla_1
typeof(mezcla_1)

mezcla_2
typeof(mezcla_2)

mezcla_3
typeof(mezcla_3)
```

## Valores faltantes (`NA`) y cuidado estadistico

`NA` representa dato faltante.
Muchas funciones estadisticas devuelven `NA` si no indicas `na.rm = TRUE`.


```r
# Vector con valor faltante.
x <- c(10, 12, NA, 15, 9)

# Sin na.rm, la media queda como NA.
mean(x)

# Con na.rm, se ignoran faltantes explicitamente.
mean(x, na.rm = TRUE)
median(x, na.rm = TRUE)
sd(x, na.rm = TRUE)

# Detecta faltantes.
is.na(x)
```

## Factores: variables categoricas en R

`factor` es muy importante para modelado y analisis de variables categoricas.
Puedes definir niveles y orden de forma explicita.


```r
# Factor sin orden.
color <- factor(c('rojo', 'azul', 'rojo', 'verde'))
color
levels(color)

# Factor ordenado (ordinal).
nivel <- factor(
  c('bajo', 'alto', 'medio', 'alto'),
  levels = c('bajo', 'medio', 'alto'),
  ordered = TRUE
)

nivel
str(nivel)
```

## Listas: estructuras heterogeneas

A diferencia de un vector atomico, una lista puede guardar objetos de distinto tipo.


```r
# Lista con tipos mezclados.
obj <- list(
  nombre = 'Ana',
  edad = 21,
  activo = TRUE,
  notas = c(9, 8, 10)
)

# Acceso por nombre y por posicion.
obj$nombre
obj[[2]]
obj[['notas']]

# Estructura completa.
str(obj)
```

## `data.frame`: tabla base para analisis de datos

`data.frame` organiza columnas (variables) y filas (observaciones).
Es fundamental en analisis estadistico en R.


```r
# Crea una tabla simple.
estudiantes <- data.frame(
  nombre = c('Ana', 'Luis', 'Marta'),
  edad = c(20, 22, 21),
  promedio = c(8.7, 9.1, 8.9),
  activo = c(TRUE, TRUE, FALSE)
)

# Explora estructura y resumen.
str(estudiantes)
summary(estudiantes)

# Seleccion de columnas y filas.
estudiantes$nombre
estudiantes[1:2, c('nombre', 'promedio')]

# Crea una nueva columna derivada.
estudiantes$aprobado <- estudiantes$promedio >= 8.0
estudiantes
```

## Conversion de tipos

Convertir tipos de forma explicita ayuda a evitar errores de interpretacion.


```r
# Conversiones frecuentes.
a <- '123'
b <- as.numeric(a)

c <- 7.9
d <- as.integer(c)

e <- c(1, 0, 1)
f <- as.logical(e)

a
b
d
f

# Cuidado: conversion invalida produce NA con warning.
as.numeric(c('10', 'abc', '30'))
```

## Reciclaje de vectores (regla importante en R)

Si operas vectores de longitudes distintas, R recicla el mas corto.
Puede ser util, pero tambien peligroso si no era tu intencion.


```r
# Ejemplo de reciclaje.
v1 <- c(1, 2, 3, 4, 5, 6)
v2 <- c(10, 20)

v1 + v2

# Recomendacion: validar longitudes antes de operar.
length(v1)
length(v2)
```

## Extra de probabilidad y estadistica

Simularemos una variable Bernoulli (0/1) y veremos como su promedio empirico estima la probabilidad.


```r
# Fija semilla para reproducibilidad.
set.seed(2026)

# Simula 2000 observaciones Bernoulli con p = 0.35.
n <- 2000
x <- sample(c(0, 1), size = n, replace = TRUE, prob = c(0.65, 0.35))

# Estadisticos descriptivos clave.
cat('p teorica = 0.35
')
cat('p estimada =', round(mean(x), 4), '
')
cat('varianza muestral =', round(var(x), 4), '
')
cat('desv. estandar =', round(sd(x), 4), '
')

# Frecuencias absoluta y relativa.
tabla <- table(x)
prop <- prop.table(tabla)
tabla
prop

# Visualizacion simple.
barplot(tabla, col = c('gray70', 'tomato'), main = 'Bernoulli simulada', xlab = 'valor', ylab = 'frecuencia')
```

## Analogias rapidas R y Python

- `vector` en R se usa mas directamente para operaciones numericas; en Python esto suele pasar via `numpy`.
- `data.frame` en R y DataFrame en Python cumplen un rol parecido en analisis tabular.
- `NA` en R requiere tratamiento explicito en estadistica, similar a `NaN` en Python.


```r
# Ejemplo comparativo de manejo de faltantes en R.
# En R base debes indicar na.rm = TRUE para omitir NA en muchos estadisticos.
datos <- c(5, 7, NA, 9)
mean(datos)                 # devuelve NA
mean(datos, na.rm = TRUE)   # calcula media ignorando NA
```

## Errores comunes de esta sesion

1. Mezclar tipos y no verificar coercion final.
2. Calcular estadisticos con `NA` sin `na.rm = TRUE`.
3. Usar factores sin controlar niveles/orden.
4. Olvidar la regla de reciclaje de vectores.


```r
# Ejemplo de error comun: asumir que suma entre vectores siempre alinea por posicion semantica.
a <- c(100, 200, 300, 400)
b <- c(1, 2)
resultado <- a + b

# Observa reciclaje de b.
a
b
resultado

# Buena practica: valida longitudes compatibles.
if (length(a) %% length(b) != 0) {
  warning('Longitudes no compatibles: posible reciclaje no deseado')
}
```
