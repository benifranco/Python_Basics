## 3) Primer repositorio local (`init`, `status`, `add`, `commit`)

## 5) Trabajar con cambios (`restore`, `rm`, `mv`)

### Teoria breve

- `git restore archivo`: descarta cambios no confirmados.
- `git rm archivo`: elimina archivo y registra el cambio.
- `git mv viejo nuevo`: renombra archivo en forma rastreable.

### Demo (terminal)

```bash

git restore notas.txt
git mv notas.txt apuntes.txt
git rm archivo_temporal.txt
git status
```

Nota: usa estos comandos con cuidado. Primero revisa estado.

## 8) Conectar repo local con GitHub

### Teoria breve

Para subir tu repo local a GitHub necesitas un remoto (`origin`).

### Demo (terminal)

```bash

# 2) En local:

git branch -M main
git push -u origin main
git remote -v
```

Despues de `-u`, puedes usar `git push` y `git pull` sin escribir rama cada vez.

## 9) Flujo diario con remoto (`fetch`, `pull`, `push`)

### Teoria breve

- `git fetch`: trae cambios remotos sin mezclar.
- `git pull`: trae y fusiona cambios.
- `git push`: sube tus commits.

### Demo (terminal)

```bash
git status
git fetch origin
git log --oneline --all --decorate
git pull origin main
git push origin main
```

Tip: cuando hay cambios de otros, primero `pull`, luego trabajas y al final `push`.
